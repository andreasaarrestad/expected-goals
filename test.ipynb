```python
import pandas as pd
from sklearn.model_selection import train_test_split
import statsmodels.formula.api as smf
from pipeline import read_xml
```

```python
data = read_xml()
data.iloc[100:120, :]
```

```python
from pipeline import transform_events


data = transform_events()
```

```python
data.columns
```

```python
on_target_ids = [30, 155]

modelchoice = "logit"

data["on_target"] = (data["type"].isin(on_target_ids)).astype(int)

data["is_away"] = 1 - data.is_home

df = data[["distance", "angle", "on_target", "dangerous_attacks_home_cum", "dangerous_attacks_away_cum", "is_home", "is_away"]]

data_train, data_test = train_test_split(df, test_size = 0.2, random_state = 43)

formula_string = "on_target ~ distance + angle + distance * angle + dangerous_attacks_home_cum * is_home + dangerous_attacks_away_cum * is_away"

if modelchoice == "linear":
    model = smf.ols(formula = formula_string, data = data_train)
    result = model.fit()
elif modelchoice == "logit":
    model = smf.logit(formula = formula_string, data = data_train)
    result = model.fit()

print(result.summary())   
```

```python
data["on_target_pred"] = result.predict(data)
data
```
